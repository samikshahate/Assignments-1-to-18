# K-Means-Clustering-

```python
## Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
#Import Dataset
crime = pd.read_csv('crime_data.csv')
crime
```

```python
crime = crime.drop(['Unnamed: 0'],axis=1)
crime
```

```python
crime.info()
```

```python
# Normalize heterogenous numerical data by using Standard Scaler
crime_norm = StandardScaler().fit_transform(crime)
```

```python
# within-cluster sum-of-squares criterion 
wcss=[]
for i in range (1,11):
    kmeans = KMeans(n_clusters=i, random_state=2)
    kmeans.fit(crime_norm)
    wcss.append(kmeans.inertia_)
```

```python
#plot K values range vs WCSS to ger Elbow graph forchoosing K (no. of clusters)
plt.plot(range(1,11),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

# Build Cluster algorithm using K=4

```python
#Cluster algorithm using K=4
cluster4 =KMeans(4,random_state=30).fit(crime_norm)
cluster4
```

```python
cluster4.labels_
```

```python
#Assign clusters to the data set
crime4 = crime.copy()
crime4['cluster4id'] = cluster4.labels_
crime4
```

```python
# Compute the centroids for K=4 clusters with 11 variables
cluster4.cluster_centers_
```

```python
#Grroup data by Cluster (K=4)
crime4.groupby('cluster4id').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize= (10,7))
plt.scatter(crime4['cluster4id'],crime4['UrbanPop'], c=cluster4.labels_)
```
